## Question 1: Multinomial logistic regression and cross validation

```python
# Read in data from data folder
strongdrink = pd.read_csv("data/strongdrink.txt")
strongdrink.columns = ['cultivar','alco','malic','ash','alk','magn','tot_phen',
                   'flav','nonfl_phen','proanth','color_int','hue','OD280rat','proline']
strongdrink.head()
```

### a) Estimate the multinomial logistic regression model

```python
# Select needed columns for the model
strongdrink = strongdrink[['cultivar', 'alco', 'malic', 'tot_phen', 'color_int']]
strongdrink.head()
```

**(1) Estimate the model on a 75% sample training set and report the two sets of estimated coefficients**

```python
X = strongdrink[['alco', 'malic', 'tot_phen', 'color_int']]
y = strongdrink[['cultivar']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .25, random_state=20)
MultLogReg = LogisticRegression(multi_class='multinomial', solver='newton-cg')
MultLogReg.fit(X_train, y_train)
#report the two sets of estimated coefficients for j = 1 and j = 2
xvar = ['alco', 'malic', 'tot_phen', 'color_int']
coef = pd.DataFrame(MultLogReg.coef_[0:2], index = ["cultivar = 1", "cultivar = 2"], columns = xvar)
coef.loc[:,"Intercept"] = MultLogReg.intercept_[:2]
print("The two sets of estimated coefficients for j = 1 and j = 2 are as follows:")
coef
```

```python
y_pred = MultLogReg.predict(X_test)
confusion_matrix = confusion_matrix(y_test, y_pred)
confusion_matrix
```

**(2) Report your error rates on the test set**

```python
print(classification_report(y_test, y_pred))
print('The error rate of class 1 is', 1 - 0.867,
      '\nThe error rate of class 2 is', 1 - 1,
      '\nThe error rate of class 3 is', 1 - 1)
print(f'Validation set MSE = {mean_squared_error(y_test, y_pred)}')
```

From the report above, one can see that cultivar 3 has the best ability at predicting. Yet, one also see from the report above that cultivar 3 has the least observations, which is conterintuitive since one might expect the one that's most accurate to be the one that has the most observations. 

### b) Perform a leave-one-out cross validation with the model from a)

```python
N = strongdrink.shape[0]
loo = LeaveOneOut()
Xvars = X.values
yvals = y.values
loo.get_n_splits(Xvars)
MSE = pd.DataFrame({'index': np.zeros(N), 
                    'error': np.zeros(N), 
                    'y_pred': np.zeros(N),
                    'y_test': np.zeros(N)})
classify = np.empty((N, 2))
for train_index, test_index in loo.split(X):
    X_train, X_test = Xvars[train_index], Xvars[test_index]
    y_train, y_test = yvals[train_index],yvals[test_index]
    LogReg = LogisticRegression(multi_class='multinomial',
                                solver='newton-cg')
    LogReg.fit(X_train, y_train)
    y_pred = LogReg.predict(X_test)
    MSE['index'][test_index] = test_index
    MSE['error'][test_index] = (y_test != y_pred).astype(int)
    MSE['y_pred'][test_index] = y_pred
    MSE['y_test'][test_index] = y_test
```

```python
error_1 = MSE[MSE['y_pred'] == 1]['error'].mean()
error_2 = MSE[MSE['y_pred'] == 2]['error'].mean()
error_3 = MSE[MSE['y_pred'] == 3]['error'].mean()
print(classification_report(MSE['y_test'], MSE['y_pred'], digits=3))
print('The error rate for class 1 is', error_1, ', std is', MSE[MSE['y_pred'] == 1]['error'].std())
print('The error rate for class 2 is', error_2, ', std is', MSE[MSE['y_pred'] == 2]['error'].std())
print('The error rate for class 3 is', error_3, ', std is', MSE[MSE['y_pred'] == 3]['error'].std())
```

From the report above, one can see that, comparing to the previous performance, here class 1's error rate 0.098 is smaller than before, 0.133. However, the error rate for class 2 and class 3 both increased from 0. This result is worse than result from (a).

```python
General_MSE = MSE['error'].mean()
print('The average MSE (for all classes) is', General_MSE, ', std is', MSE['error'].std())
```

Comparing to 0.0454 which is the previous average MSE, the new average MSE is larger. 

### c) Perform a k-fold cross validation

```python
k = 4

kf = KFold(n_splits=k, random_state=10, shuffle=True)
kf.get_n_splits(Xvars)
MSE_vec_kf = np.zeros(k)
ytests = np.array([])
ypreds = np.array([])
k_ind = int(0)

for train_index, test_index in kf.split(Xvars):
    print('k index=', k_ind)
    X_train, X_test = Xvars[train_index], Xvars[test_index]
    y_train, y_test = yvals[train_index], yvals[test_index]
    ytests = np.append(ytests, y_test)
    LogReg = LogisticRegression(multi_class='multinomial', solver='newton-cg')
    LogReg.fit(X_train, y_train)
    y_pred = LogReg.predict(X_test)
    ypreds = np.append(ypreds, y_pred)
    mse_accu = 0
    for i in range(len(y_test)):
        if y_test[i] == y_pred[i]:
            indicator = 1
        else:
            indicator = 0
        mse_accu += 1 - indicator
    MSE_vec_kf[k_ind] = mse_accu / len(y_test)
    print('MSE for test set', k_ind, ' is', MSE_vec_kf[k_ind])
    print(classification_report(y_test, y_pred))
    k_ind += 1   
```

From the print statement above, one can see that the average error rate for class 1 is smaller than in part(a) and larger for class 2 and class 3 in part (a). Performing K-fold has led to a similar result as the leave-one-out cross-validation. The result, comparing to part (a), has becoming worse since in part (a) both class 2 and class 3 has an error rate of 0. 

**(2) Report the k-fold estimate for the test MSE as the average MSE**

```python
MSE_kf = MSE_vec_kf.mean()
MSE_kf_std = MSE_vec_kf.std()
print('test estimate MSE k-fold=', MSE_kf,
      'test estimate MSE standard err=', MSE_kf_std)
```

## Question 2 : Splines and Interpolation

```python
# read in data
coolindex = pd.read_csv("data/CoolIndex.txt",  names = ['age', "coolindex"], header = None,
                        dtype = {0:np.int32, 1:np.float64})
coolindex.head()
```

### a) create a scatterplot

```python
fig, ax = plt.subplots()
ax.scatter(coolindex['age'], coolindex['coolindex'], c = "green")
ax.set_xlabel("Age", fontsize = 12)
ax.set_ylabel("Coolness Index", fontsize = 12)
ax.set_title("Scatter plot of Coolness Index aqainst Age", fontsize = 15)
```

### b) Use OLS regression to fit a stepwise function to these data

```python
age = coolindex["age"]
CI = coolindex["coolindex"]
age_11_22 = ((age >= 11) & (age < 22)).astype(int)
age_22_40 = ((age >= 22) & (age < 40)).astype(int)
age_40_59 = ((age >= 40) & (age < 59)).astype(int)
age_59_77 = ((age >= 59) & (age < 77)).astype(int)
age_77_95 = ((age >= 77) & (age <= 95)).astype(int)
X_step = pd.DataFrame(dict(age_11_22=age_11_22, age_22_40=age_22_40,
                           age_40_59=age_40_59, age_59_77=age_59_77,
                           age_77_95=age_77_95))
```

```python
reg = sm.OLS(endog=CI, exog=X_step, missing='drop')
reg_results = reg.fit()
y_pred_1 = reg_results.predict(X_step)
```

**(1) plot the stepwise function**

```python
ax.plot(age, y_pred_1, color = 'red', label ="step function")
ax.set_title("Stepwise Function and Scatterplot of Coolness Index against Age",fontsize = 15)
ax.legend()

fig
```

**(2) Report your estimated step function values for each bin**

```python
# report beta values
for i in range(5):
    print("Beta_{} = {}".format(i+1, reg_results.params[i]))
```

**(3) The predicted coolness of a 73-year old from the step function**

```python
print("The predicted coolness of a 73-year old is")
reg_results.params[3]
```

### c) Fit a linear spline to the data over the 5 age bins from b) 
**(1) plot linear spline**

```python
knots = [22, 40, 59, 77]
spl_ci = LSQUnivariateSpline(age.values, coolindex.coolindex.values, knots, k=1)
age_vec = np.linspace(coolindex.age.values.min(), coolindex.age.values.max(), 1000)
ax.plot(age_vec, spl_ci(age_vec), label = "linear spline")
ax.legend()
ax.set_title("Linear Spline, Stepwise Function, and Scatterplot of Coolness Index against Age", fontsize = 15)

fig
```

**(2) the predicted coolness of a 73-year old from the linear spline**

```python
print("The predicted coolness of a 73-year old from linear spline is")
print(spl_ci(73))
```

### d) Fit a cubic spline to the data over the 5 age bins from b)
**(1)plot the cubic spline**

```python
cubspl_ci = LSQUnivariateSpline(coolindex.age.values, coolindex.coolindex.values, knots, k=3)
age_vec = np.linspace(coolindex.age.values.min(), coolindex.age.values.max(), 1000)
ax.plot(age_vec, cubspl_ci(age_vec), label = "cubic spline")
ax.legend()
ax.set_title("Cubic Spline, Linear Spline, Stepwise Function, and Scatterplot of Coolness Index against Age", fontsize = 14)

fig
```

**(2) the predicted coolness of a 73-year old from the cubic spline**

```python
print("The predicted coolness of a 73-year old from cubic spline is")
print(cubspl_ci(73))
```
